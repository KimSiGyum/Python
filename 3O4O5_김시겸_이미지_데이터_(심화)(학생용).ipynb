Ïù¥ÎØ∏ÏßÄÎäî ÏàòÏùò Î∞∞Ïó¥Ïù∏ ÌîΩÏÖÄ(pixel)Î°ú ÌëúÌòÑÎêòÍ∏∞ ÎïåÎ¨∏Ïóê <u>ÏàòÌïôÏ†Å Î∞©Î≤ïÏùÑ Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º Ï≤òÎ¶¨</u>Ìï† Ïàò ÏûàÎã§.

**‚ñ∑ <u>ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏</u>**
>Ïã§ÏäµÏùÑ ÏúÑÌï¥ ÏïÑÎûò ÏΩîÎìúÎ•º Ïö∞ÏÑ† Ïã§ÌñâÌïòÏûê. (‚Äª OpenCV Ï∞∏Í≥† : https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html)

```python
import numpy as np # ÎÑòÌååÏù¥ Ìå®ÌÇ§ÏßÄ Í∞ÄÏ†∏Ïò§Í∏∞
import cv2 # OpenCV ÌååÏù¥Ïç¨ ÎùºÏù¥Î∏åÎü¨Î¶¨ Í∞ÄÏ†∏Ïò§Í∏∞
from google.colab.patches import cv2_imshow # ÏΩîÎû©ÏóêÏÑú ÏõπÎ∏åÎùºÏö∞Ï†ÄÏóê Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•ÌïòÍ∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©
```

---
####**Ïù¥ÎØ∏ÏßÄ ÌÅ¨Î°úÌïë (cropping)**
- ÌîΩÏÖÄÏùò ÏúÑÏπòÎ•º Î∞îÍøî Ïù¥ÎØ∏ÏßÄÏùò ÌòïÌÉúÎ•º Î≥ÄÌòïÌïòÎäî Í∏∞ÌïòÌïôÏ†Å Î≥ÄÌôòÏùò ÌïòÎÇò
- Ïù¥ÎØ∏ÏßÄÏùò ÏùºÎ∂ÄÎ•º ÏûòÎùºÎÇ¥Îäî Í≤É

**‚ñ∑ <u>Ïã§ÏäµÌï† Ïù¥ÎØ∏ÏßÄ Ï§ÄÎπÑÌïòÍ∏∞</u>**
>ÏΩîÎû© ÏÑúÎ≤ÑÏóê Ìè¨ÏºìÎ™¨ Í∑∏Î¶ºÏùÑ ÏóÖÎ°úÎìúÌïòÍ≥† OpenCVÎ•º Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏñ¥Ïò§Ïûê.<br>

```python
img = cv2.imread('pikachu.png') # Ïù¥ÎØ∏ÏßÄ ÏùΩÏñ¥Ïò§Í∏∞
cv2_imshow(img)
```

```python
img.shape # Ïù¥ÎØ∏ÏßÄÏùò ÌòïÌÉú ÌôïÏù∏ÌïòÍ∏∞
```

**‚ñ∑ <u>Ïù¥ÎØ∏ÏßÄ ÏûêÎ•¥Í∏∞</u>**
>Ïù¥ÎØ∏ÏßÄ(img)Ïùò Ïò§Î•∏Ï™Ω ÏúÑ 1/4 Ï°∞Í∞ÅÏùÑ ÏûòÎùºÎÇ¥Ïñ¥ Ï∂úÎ†•Ìï¥Î≥¥Ïûê.

```python
crop = img[:img.shape[0]//2, img.shape[1]//2:] # (293, 383, 3)
cv2_imshow(crop)
```

---
####**Ïù¥ÎØ∏ÏßÄ ÌîåÎ¶¨Ìïë (flipping)**
- ÌîΩÏÖÄÏùò ÏúÑÏπòÎ•º Î∞îÍøî Ïù¥ÎØ∏ÏßÄÏùò ÌòïÌÉúÎ•º Î≥ÄÌòïÌïòÎäî Í∏∞ÌïòÌïôÏ†Å Î≥ÄÌôòÏùò ÌïòÎÇò
- Ï¢åÏö∞ Î∞òÏ†Ñ, ÏÉÅÌïò Î∞òÏ†Ñ Îì±

**‚ñ∑ <u>Ïã§ÏäµÌï† Ïù¥ÎØ∏ÏßÄ Ï§ÄÎπÑÌïòÍ∏∞</u>**
>ÏΩîÎû© ÏÑúÎ≤ÑÏóê three.pngÎ•º ÏóÖÎ°úÎìúÌïòÍ≥† OpenCVÎ•º Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏñ¥Ïò§Ïûê.<br>
‚Äª Í∞ÑÌé∏Ìï®ÏùÑ ÏúÑÌï¥ ÌöåÏÉâÏ°∞(grayscale) Ïù¥ÎØ∏ÏßÄÎ°ú Ïã§Ïäµ

```python
img = cv2.imread('three.png', cv2.IMREAD_GRAYSCALE) # Ïù¥ÎØ∏ÏßÄÎ•º ÌöåÏÉâÏ°∞Î°ú ÏùΩÏñ¥Ïò§Í∏∞
cv2_imshow(img)
```

```python
img.shape # Ïù¥ÎØ∏ÏßÄÏùò ÌòïÌÉú ÌôïÏù∏ÌïòÍ∏∞
```

**‚ñ∑ <u>Ïù¥ÎØ∏ÏßÄ Ï¢åÏö∞ Î∞òÏ†ÑÌïòÍ∏∞</u>**
>Ïù¥ÎØ∏ÏßÄ(img)Î•º Ï¢åÏö∞ Î∞òÏ†ÑÌïòÏó¨ Ï∂úÎ†•Ìï¥Î≥¥Ïûê.<br>
**<font color='tomato'>{ ÌûåÌä∏ }</font>** Ïñ¥ÎñªÍ≤å ÌïòÎ©¥ Î™®Îì† ÌîΩÏÖÄÏùò ÏúÑÏπòÎ•º Ï¢åÏö∞Î°ú Î≥ÄÍ≤ΩÌï† Ïàò ÏûàÏùÑÍπåÏöî?

```python
mirror = np.zeros((28, 28))
for i in range(28):
  for j in range(28):
    mirror[i, j] = img[i, 27-j]
cv2_imshow(mirror)
```

```python
mirror = np.flip(img, axis=1) # ÎÑòÌååÏù¥ flip() Ìï®Ïàò Ïù¥Ïö©
cv2_imshow(mirror)
```

```python
mirror = img[:,::-1] # ÎÑòÌååÏù¥ Î∞∞Ïó¥Ïùò Í∏∞Îä• Ïù¥Ïö©
cv2_imshow(mirror)
```

**üëç<font color='dodgerblue'>{ ÎèÑÏ†ÑÍ≥ºÏ†ú }</font>** Ïù¥ÎØ∏ÏßÄ(img)Î•º ÏÉÅÌïò Î∞òÏ†ÑÌïòÏó¨ Ï∂úÎ†•Ìï¥Î≥¥Ïûê.

```python
mirror = np.zeros((28, 28))
for i in range(28):
    mirror[i] = img[27-i]

cv2_imshow(mirror)
```

```python
mirror = np.flip(img, axis=0)# ÎÑòÌååÏù¥ flip() Ìï®Ïàò Ïù¥Ïö©
cv2_imshow(mirror)
```

```python
mirror = img[::-1, :]# ÎÑòÌååÏù¥ Î∞∞Ïó¥Ïùò Í∏∞Îä• Ïù¥Ïö©
cv2_imshow(mirror)
```

**üëç<font color='dodgerblue'>{ ÎèÑÏ†ÑÍ≥ºÏ†ú }</font>** Ï¢ãÏïÑÌïòÎäî Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º ÏóÖÎ°úÎìúÌïòÏó¨ Ï¢åÏö∞Î•º Î∞òÏ†ÑÏãúÏºúÎ≥¥Ïûê.

```python
cimg = cv2.imread('dog.png')
cv2_imshow(cimg)
```

```python
cimg.shape
```

```python
mirror = np.zeros((cimg.shape[0], cimg.shape[1], 3))
for i in range(cimg.shape[0]):
    for j in range(cimg.shape[1]):
        mirror[i, j] = cimg[i, cimg.shape[1]-1-j]


cv2_imshow(mirror)
```

```python
mirror = np.flip(cimg, axis=1)# ÎÑòÌååÏù¥ flip() Ìï®Ïàò Ïù¥Ïö©
cv2_imshow(mirror)
```

```python
mirror = cimg[:,::-1] # ÎÑòÌååÏù¥ Î∞∞Ïó¥Ïùò Í∏∞Îä• Ïù¥Ïö©
cv2_imshow(mirror)
```

- ÌîΩÏÖÄÏùò RGB Í∞íÏùÑ Î≥ÄÍ≤ΩÌï¥ ÌäπÏ†ï ÏúÑÏπòÏùò ÏÉâÏÉÅÏùÑ Î≥ÄÍ≤ΩÌï† Ïàò ÏûàÏùå (OpenCVÏóêÏÑú ÌîΩÏÖÄÏùÄ BGR Ïàú)

**‚ñ∑ <u>Ïã§ÏäµÌï† Ïù¥ÎØ∏ÏßÄ Ï§ÄÎπÑÌïòÍ∏∞</u>**
>ÏΩîÎû© ÏÑúÎ≤ÑÏóê dog.pngÎ•º ÏóÖÎ°úÎìúÌïòÍ≥† OpenCVÎ•º Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏñ¥Ïò§Ïûê.

```python
img = cv2.imread('dog.png')
cv2_imshow(img)
```

```python
img.shape
```

**‚ñ∑ <u>Ïù¥ÎØ∏ÏßÄ ÏÉâÏπ†ÌïòÍ∏∞</u>**
> Îã§Ïùå ÏΩîÎìúÏóê ÎåÄÌïú ÏÑ§Î™ÖÏùÑ Ï£ºÏÑùÏúºÎ°ú Ï†ÅÏñ¥Î≥¥Ïûê.

```python
img[:150, :50] = [0, 0, 0]
cv2_imshow(img)
```

**üëç<font color='dodgerblue'>{ ÎèÑÏ†ÑÍ≥ºÏ†ú }</font>** Ïù¥ÎØ∏ÏßÄ(img)ÏóêÏÑú Ïò§Î•∏Ï™Ω ÏïÑÎûò 50 * 50 ÏòÅÏó≠Ïùò ÏÉâÏÉÅÏùÑ Î∞îÍøîÎ≥¥Ïûê. (‚Äª ÏÉâÏÉÅÏΩîÎìú Ï∞∏Í≥† : https://en.wikipedia.org/wiki/Web_colors)

```python
img[-50:, -50:] = [0, 255, 255]
cv2_imshow(img)
```

**‚ñ∑ <u>Ïù¥ÎØ∏ÏßÄ ÏÉâ Î≥ÄÏ°∞ÌïòÍ∏∞</u>**
> Îã§Ïùå ÏΩîÎìúÏùò Ïã§ÌñâÍ≤∞Í≥ºÎ•º ÏòàÏÉÅÌï¥Î≥¥Ïûê.

```python
cv2_imshow(img * 0.5) # Î™®Îì† ÌîΩÏÖÄ Í∞íÏùÑ Î∞òÏúºÎ°ú
```

> Îã§Ïùå ÏΩîÎìúÏùò Ïã§ÌñâÍ≤∞Í≥ºÎ•º ÏòàÏÉÅÌï¥Î≥¥Ïûê.

---
####**Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ ÏîåÏö∞Í∏∞**
- ÌïÑÌÑ∞(filter)Îäî Ïù¥ÎØ∏ÏßÄÏóê ÌäπÏ†ï Ìö®Í≥ºÎ•º Ï£ºÍ±∞ÎÇò ÌäπÏßïÏùÑ Í∞ïÌôîÌïòÍ∏∞ ÏúÑÌï¥ Ï†ÅÏö©ÌïòÎäî ÏàòÌïôÏ†Å Ïó∞ÏÇ∞
- Ï£ºÎ≥Ä ÌîΩÏÖÄÏùò Ï†ïÎ≥¥Î•º Ï¢ÖÌï©ÌïòÏó¨ ÏÉàÎ°úÏö¥ ÌîΩÏÖÄ Í∞íÏùÑ Í≥ÑÏÇ∞Ìï®
- Îã§ÏñëÌïú ÌïÑÌÑ∞ : https://setosa.io/ev/image-kernels/

**‚ñ∑ <u>Ïã§ÏäµÌï† Ïù¥ÎØ∏ÏßÄ Ï§ÄÎπÑÌïòÍ∏∞</u>**
>ÏΩîÎû© ÏÑúÎ≤ÑÏóê shape.pngÎ•º ÏóÖÎ°úÎìúÌïòÍ≥† OpenCVÎ•º Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏñ¥Ïò§Ïûê.

```python
img = cv2.imread('shape.png')
cv2_imshow(img)
```

**‚ñ∑ <u>ÌïÑÌÑ∞Î°ú Ï†ÅÏö©Ìï¥Î≥¥Í∏∞</u>**
>Îã§Ïùå ÏΩîÎìúÏùò Ïã§ÌñâÍ≤∞Í≥ºÎ•º ÌôïÏù∏Ìï¥Î≥¥Í≥† ÌïÑÌÑ∞Î•º Ï†ÅÏö©Ìïú Í≤∞Í≥ºÍ∞Ä Ïôú Í∑∏Î†áÍ≤å ÎÇòÏôîÎäîÏßÄ ÏÉùÍ∞ÅÌï¥Î≥¥Ïûê.

```python
filter = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])
filtered = cv2.filter2D(img, -1, filter)
cv2_imshow(filtered)
```

**üëç<font color='dodgerblue'>{ ÎèÑÏ†ÑÍ≥ºÏ†ú }</font>** ÎßàÏùåÏóê ÎìúÎäî Ïù¥ÎØ∏ÏßÄÎ•º ÏóÖÎ°úÎìúÌïòÍ≥† Îã§ÏñëÌïú ÌïÑÌÑ∞Î•º Ï†ÅÏö©Ìï¥Î≥¥Ïûê.
- Î∏îÎü¨(Blur) / ÏÑ†Î™ÖÌïòÍ≤å(Sharpen) / Ïú§Í≥ΩÏÑ†(Outline)

```python
img = cv2.imread('dog.png')
cv2_imshow(img)
```

```python
filter = np.ones((5,5), np.float32)/ 25
filtered = cv2.filter2D(img, -1, filter)
cv2_imshow(filtered)
```

**üëç<font color='dodgerblue'>{ ÎèÑÏ†ÑÍ≥ºÏ†ú }</font>** ÎÇòÎßåÏùò ÌïÑÌÑ∞Î•º ÎßåÎì§Ïñ¥Î≥¥Ïûê.

```python
filter = np.array([[0, 0, 0],
                   [0, 1, 0],
                   [0, 0, 0]]) # Ïò§Î¶¨ÏßÄÎÑê
```

```python
filter = np.array([[1, 0, -1],
                   [0, 0, 0],
                   [-1, 0, 1]])
```

```python
filter = np.array([[30, 1, 0],
                   [-1, 0, 0],
                   [1, 0, -30]])
```

```python
filter = np.array([[-1,-1,-1],
                   [-1,8,-1],
                   [-1,-1,-1]])
```

```python
filtered = cv2.filter2D(img, -1, filter)
cv2_imshow(filtered)
```
